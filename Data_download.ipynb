```python
import pandas as pd
df = pd.read_csv(f"{CIRCOR_ROOT}/training_data.csv")
print(df.shape)
df.head()
```

```python
df["Recording locations:"].unique()
```

```python
sites = ["AV","PV","TV","MV","Phc"]

def parse_sites(s):
    if pd.isna(s): return []
    return [t.strip() for t in str(s).split("+")]

df["sites_list"] = df["Recording locations:"].apply(parse_sites)
df["n_recordings_listed"] = df["sites_list"].str.len()

# Booleans for presence of each site (ignores duplicates)
for site in sites:
    df[f"has_{site}"] = df["sites_list"].apply(lambda L: site in L)

# 1) How variable is coverage?
coverage = df["n_recordings_listed"].value_counts().sort_index()

# 2) Does outcome correlate with site coverage? (watch for confounding)
by_outcome_n = df.groupby("Outcome")["n_recordings_listed"].describe()
by_outcome_sites = df.groupby("Outcome")[[f"has_{s}" for s in sites]].mean()
```

```python
print(coverage)
print(by_outcome_n)
print(by_outcome_sites)
```

```python
SITE_PRIORITY = ["MV", "AV", "PV", "TV", "Phc"]  # availability-descending
USE_MOST_AUDIBLE = True  # set False to disable the override globally
```

```python
import os, re, json, time
from pathlib import Path
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from collections import Counter
```

```python
TRAIN_PCT, VAL_PCT, TEST_PCT = 0.70, 0.15, 0.15
SPLIT_SEED = 42
WAV_DIR = Path(CIRCOR_ROOT) / "training_data"   # expects *.wav here
OUT_DIR = Path(CIRCOR_ROOT) / "splits"
OUT_DIR.mkdir(parents=True, exist_ok=True)
```

```python
patient_df = df.copy()
```

```python
patient_df = patient_df.rename(columns={"Patient ID": "patient_id"})
patient_df["patient_id"] = patient_df["patient_id"].astype(str).str.strip()
patient_df["Outcome"] = patient_df["Outcome"].astype(str).str.strip()
```

```python
patient_df = patient_df[patient_df["Outcome"].isin(["Normal", "Abnormal"])].reset_index(drop=True)
```

```python
y = patient_df["Outcome"]
all_ids = patient_df["patient_id"]

train_ids, temp_ids = train_test_split(
    all_ids, test_size=(1 - TRAIN_PCT), stratify=y, random_state=SPLIT_SEED
)
temp_df = patient_df[patient_df["patient_id"].isin(temp_ids)]
val_ids, test_ids = train_test_split(
    temp_df["patient_id"],
    test_size=TEST_PCT / (TEST_PCT + VAL_PCT),
    stratify=temp_df["Outcome"],
    random_state=SPLIT_SEED,
)

train_ids = set(map(str, train_ids))
val_ids   = set(map(str, val_ids))
test_ids  = set(map(str, test_ids))
```

```python
assert train_ids.isdisjoint(val_ids) and train_ids.isdisjoint(test_ids) and val_ids.isdisjoint(test_ids), "Patient overlap across splits!"
def label_counts(ids):
    lab = patient_df[patient_df["patient_id"].isin(ids)]["Outcome"]
    return dict(Counter(lab))
print("[Patients]  train:", len(train_ids), label_counts(train_ids))
print("[Patients]  val  :", len(val_ids),   label_counts(val_ids))
print("[Patients]  test :", len(test_ids),  label_counts(test_ids))
```

```python
wav_pat = re.compile(r"^(\d+)_([A-Za-z]+).*\.wav$")
pid_to_files = {}
for p in WAV_DIR.glob("*.wav"):
    m = wav_pat.match(p.name)
    if not m:
        continue
    pid, site = m.group(1), m.group(2)
    pid_to_files.setdefault(pid, []).append((p, site))
```

```python
wav_pat = re.compile(r"^(\d+)_([A-Za-z]+).*\.wav$")
pid_to_files = {}
for p in WAV_DIR.glob("*.wav"):
    m = wav_pat.match(p.name)
    if not m:
        continue
    pid, site = m.group(1), m.group(2)
    pid_to_files.setdefault(pid, []).append((p, site))

# ---- Helper: build a manifest DataFrame for a set of patient IDs ----
def build_manifest(pid_set: set, name: str) -> pd.DataFrame:
    rows = []
    sub = patient_df[patient_df["patient_id"].isin(pid_set)].set_index("patient_id")
    missing_patients = []
    for pid in sorted(pid_set):
        files = pid_to_files.get(pid, [])
        if not files:
            missing_patients.append(pid)
            continue
        outcome = sub.loc[pid, "Outcome"]
        murmur  = sub.loc[pid, "Murmur"] if "Murmur" in sub.columns else np.nan
        for p, site in files:
            rows.append({
                "audio": str(p),              # absolute path to WAV
                "patient_id": pid,
                "site": site,                 # AV/PV/TV/MV/Phc
                "outcome": outcome,           # Normal/Abnormal
                "murmur": murmur              # Present/Absent/Unknown (if available)
            })
    dfm = pd.DataFrame(rows)
    if missing_patients:
        print(f"[{name}] Patients with no WAV files found: {len(missing_patients)} → (e.g.) {missing_patients[:5]}")
    return dfm
```

```python
train_manifest = build_manifest(train_ids, "train")
val_manifest   = build_manifest(val_ids,   "val")
test_manifest  = build_manifest(test_ids,  "test")
```

```python
ts = time.strftime("%Y%m%d_%H%M%S")
split_dir = OUT_DIR / f"patient_splits_{ts}"
split_dir.mkdir(parents=True, exist_ok=True)
```

```python
train_manifest.to_csv(split_dir / "train_manifest.csv", index=False)
val_manifest.to_csv(split_dir / "val_manifest.csv", index=False)
test_manifest.to_csv(split_dir / "test_manifest.csv", index=False)
```

```python
# save patient ID lists
pd.Series(sorted(train_ids)).to_csv(split_dir / "train_patients.txt", index=False, header=False)
pd.Series(sorted(val_ids)).to_csv(split_dir / "val_patients.txt", index=False, header=False)
pd.Series(sorted(test_ids)).to_csv(split_dir / "test_patients.txt", index=False, header=False)
```

```python
def summarize(dfm, name):
    by_outcome = dfm["outcome"].value_counts().to_dict()
    print(f"[{name}] patients={df[ df['Patient ID'].astype(str).isin(set(dfm['patient_id'])) ].shape[0]:4d} | "
          f"wavs={len(dfm):5d} | outcome={by_outcome}")

summarize(train_manifest, "train")
summarize(val_manifest,   "val")
summarize(test_manifest,  "test")

print(f"\nWrote manifests to: {split_dir}")
```
